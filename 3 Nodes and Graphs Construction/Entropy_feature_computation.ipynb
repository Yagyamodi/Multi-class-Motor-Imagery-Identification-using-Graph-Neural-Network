```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import os  
import math
import statistics
from scipy.stats import kurtosis
import antropy
import pyeeg
import eeglib
import scipy as sp
import ordpy
```

```python

s = "A01T_1";   #read file name
readFile = "Segmentation/A01/Training/" + s  + "/1.csv"  #read file location
saveDirectory = "Graphs/A01/Training/"  #save location
training = pd.read_csv(readFile)
#makeFolder(s,saveDirectory)
print(training.shape[1])
n = training.shape[1]
print(training.shape)
x=np.array(training)
print(x.shape)
for i in range(0,22):
  X=x.T[i]
  labelling = [0] * 22

  #print("data ", x)
  s=np.std(X)
  labelling[0] = pyeeg.ap_entropy(X, 2, 0.2*s)
  #Sample Entropy : s(X,m,r) : Generally we take the value of m to be 2 and the value of r to be 0.2*std.
  #the above also applies for ap_entropy  
  print("\nap entropy: ", labelling[0], "\n--------------")
  
  labelling[1] = antropy.sample_entropy(X) 
  print("simple entropy: ", labelling[1], "\n--------------")

  labelling[2] = antropy.svd_entropy(X, 5, 2, normalize=True)
  print("svd (using entropy library) entropy: ", labelling[2], "\n--------------")
  #labelling[2] = pyeeg.svd_entropy(X, 5, 2)
  #print("svd entropy: ", labelling[2], "\n--------------")
  
  labelling[3] = antropy.perm_entropy(X, 5, 2, normalize=True)
  print("permutation entropy: ", labelling[3], "\n--------------")
  
  labelling[4] = antropy.spectral_entropy(X, sf=250, method='welch', normalize=True) # sf is sampling frequency
  print("spectral entropy: ", labelling[4], "\n--------------")

  labelling[5] = antropy.higuchi_fd(X);
  print("higuchi fractal dimensino", labelling[5], "\n-------------")

  labelling[6] = pyeeg.fisher_info(X, 5, 2)
  print("fisher info", labelling[6], "\n-------------")

  D = pyeeg.first_order_diff(X)
  labelling[7] = pyeeg.hjorth(X, D)
  print("hjorth parameters", labelling[7], "\n-------------")
```
