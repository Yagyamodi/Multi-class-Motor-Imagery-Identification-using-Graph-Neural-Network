```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import os  
import hfda
import math
import statistics

    
def makeFolder(directory,parent_dir):
      # os.makedirs() method will raise  
      # an OSError if the directory  
      # to be created already exists  
          
      # Path  
      path = os.path.join(parent_dir, directory)  
          
      # Create the directory  
      # 'Nikhil'  
      os.makedirs(path)  
      print("Directory '% s' created" % directory) 
      return 1

def HiguchiFractal(X, k_max):
    print("hi")
    D = hfda.measure(X, k_max)
    print("hii")
    return D

def KatzFractal(X):
    d = -100000
    n = X.shape[0]
    L = 0.000000
    for i in range(0,n-2):
      tmp1 = (X.loc[i+1]-X.loc[i])*(X.loc[i+1]-X.loc[i]) + 0.4*0.4;
      tmp1 = math.sqrt(tmp1)
      L += tmp1
      tmp2 = math.sqrt((X.loc[i+1]-X.loc[0])*(X.loc[i+1]-X.loc[0]) + (0.4*i)*(0.4*i))    
      d = max(d,tmp2)
    
    print("d: ", d, " ,L: ", L)
    a = math.log(n)
    D = a/(a+math.log(d/L))

    print("Katz Fractal ", D);
    print("-------------")
    return D

def PetrosianFractal(X):
    mean = 0.000000
    n = X.shape[0]
    for i in range(0,n):
        mean += X.loc[i];

    mean = mean/n
    print("mean: ", mean)

    z = [[0] * 1] * n   # n rows, 1 column, all equal to zero
    for i in range(0,n):
        if (X.loc[i] > mean).bool():
          z[i] = 1
        else:
          z[i] = -1

    nDel = 0;
    for i  in range(0,n-1):
        nDel += abs(z[i+1] - z[i])

    nDel = (1.00*nDel)/2
    a = math.log10(n)
    D = a/(a + math.log10(n/(n+0.4*nDel)))
    print("Petrosian Fractal ", D)
    print("------------------")
    return D

def SevcikFractal(X):
  n = X.shape[0]
  x_min = 100000000
  y_min = 0
  x_max = -10000000000
  y_max = 0.4*(n-1)

  xStar = [0]*n
  yStar = [0]*n

  for i in range(0,n):
    if (X.loc[i] > x_max).bool():
      x_max = X.loc[i]
    if (X.loc[i] < x_min).bool():
      x_min = X.loc[i]

  for i in range(0,n):
    xStar[i] = (X.loc[i] - x_min)/(x_max - x_min)
    yStar[i] = (0.4*i - y_min)/(y_max - y_min)

  L = 0
  for i in range(0,n-1):
    tmp1 = math.sqrt((yStar[i+1]-yStar[i])**2 + (xStar[i+1]-xStar[i])**2)
    L += tmp1

  D = 1 + (math.log(L) + math.log(2))/math.log(2*(n-1))
  print("Sevcik Fractal ", D)
  print("----------------")
  return D

def findSquareMean(X):
  n = X.shape[0]
  mean = 0;
  for i in range(0,n):
    mean += X.loc[i]**2
  
  mean = mean/n
  return mean

def HajorthParameters(xV):
    n = xV.shape[0];
    dxV = np.diff([xV]);
    ddxV = np.diff([dxV]);
    mx2 = findSquareMean(xV);
    mdx2 = findSquareMean(dxV);
    mddx2 = findSquareMean(ddxV);

    mob = mdx2 / mx2;
    complexity = math.sqrt(mddx2 / mdx2 - mob);
    mobility = math.sqrt(mob);

    print("Hajorth parameters: mobility = ", mobility, ", complexity = ", complexity);
    print("------------------")
    return {mobility, complexity};
  

s = "A01T_1";   #read file name
readFile = "Segmentation/A01/Training/" + s  + "/1.csv"  #read file location
saveDirectory = "Graphs/A01/Training/"  #save location
training = pd.read_csv(readFile)
#makeFolder(s,saveDirectory)
print(training.shape[1])
n = training.shape[1]

for node_no in range (0,1):
    labelling = [0] * 22
    i = 0
    # Get Kth Column of Matrix
    x = training.iloc [ : , node_no :node_no+1 ]
    #y = HiguchiFractal(x,4)
    #print("data ", x)
    labelling[i] = KatzFractal(x)
    i = i+1
    labelling[i] = PetrosianFractal(x)
    i = i+1;
    labelling[i] = SevcikFractal(x)
    i = i+1;
    labelling[i],labelling[i+1] = HajorthParameters(x)
    i += 2
    

    
```
